```python
import numpy as np
from matplotlib import pyplot as plt

# You can also shift the scale by generating the number using np.random.rand
l0 = np.random.rand(80,2)
l0 = 1+l0*14
l1 = np.random.rand(80,2)
l1 = 14+l1*20
l1 = np.concatenate((l0, l1))
l2 = np.array([0 for _ in range(80)]+[1 for _ in range(80)])
print(l1.shape)
print(l2.shape)
# We have our data generated
plt.scatter(l1[:,0],l1[:,-1], c = l2)
plt.show()
```
